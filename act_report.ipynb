<h1><center> We Rate Dogs (@DogRates) : Insights & Visualizations </center></h1>

### This report revolves around the analysis of more than 2000 tweets on a twitter account ‘WeRateDogs’ (@dog_rates) which rates people's dogs with a humorous comments about the dog.

I started off with finding the most widely used application to post the tweets.

I plotted a bar graph that shows the no. of tweets posted via each source and it clearly shows that most tweets has been posted via Twitter for iPhone.

Then, I was curious to know what could be the most popular dog name in all these tweets, I though why not find the 10 most common dog names?

Upon plotting this data, I found out that Cooper, Lucy and Charlie are the most common dog names which have 10 others as the same name as them, followed by Oliver and Tucker.

Majority of the tweets were about Golden Retrievers which makes it the most popular dog on DogRates, followed by Labrador Retriever and Pembroke.

Then I went ahead to find the ost common dog stage using a pie chart, For 85% of the dogs, we didn't had the dog stage in the tweet information. Out of the remaining 15% dogs, most of the dogs were in pupper stage (10%), followed by duggo (3.2%), puppo (~1% )and floofer (<1%).
